```python
import zmq
import msgpack
from PIL import Image
import io
import matplotlib.pyplot as plt
ctx = zmq.Context()
sock = ctx.socket(zmq.SUB)
sock.connect("tcp://localhost:5555")
sock.setsockopt_string(zmq.SUBSCRIBE, "")

while True:
    print("Waiting")
    raw = sock.recv()
    print("Rec")
    obj = msgpack.unpackb(raw, raw=False) # raw=False handles string decoding

    # Wrap bytes back into an Image
    image = Image.open(io.BytesIO(obj['data']))
    # Show using matplotlib
    plt.imshow(image)
    plt.title("Received Frame")
    plt.axis("off")   # Hide axis
    plt.show()
```
